```python
import imageio
import numpy as np
import scipy
from matplotlib import pyplot as plt
img = imageio.v2.imread('heart.jpg')
# Create a vector of points representing the coordinates and the pixel values for each pixel in the image
coordinates = np.indices(img.shape[:2]).reshape(2, -1).T
pixel_values = img.reshape(-1, img.shape[2])
points = np.hstack((coordinates, pixel_values))
points = points[::50]
points = points[np.all(points[:, 2:] != 255, axis=1)]


# Add random noise to the points
noise = np.random.normal(0, 10, points.shape)
noisy_points = points + noise
print(noisy_points)

# plot the image contents read from imageio
# Display the image
plt.figure(figsize=(10, 8))
plt.scatter(noisy_points[:, 0], noisy_points[:, 1], c='red', alpha=0.5)
plt.title('Heart Image')
plt.show()
```
