```python
import numpy as np
import matplotlib.pyplot as plt
labelStrings = []
with open("./iris.csv") as f:
    D = []
    L = []
    for line in f.readlines():
        fields = line.split(',')
        features = list(map(float, fields[0:4]))
        if not fields[4] in labelStrings:
            labelStrings.append(fields[4])
        L.append(labelStrings.index(fields[4]))
        D.append(features)

    D = np.array(D).T
    L = np.array(L)
    print(D)
    print(L)

    D0 = D[:, (L == 0)]
    D1 = D[:, (L == 1)]
    D2 = D[:, (L == 2)]
    plt.hist(D0[0], density=True, alpha = 0.5)
    plt.hist(D1[0], density=True, alpha = 0.5)
    plt.hist(D2[0], density=True, alpha = 0.5)
```
