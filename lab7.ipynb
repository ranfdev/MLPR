```python
import numpy as np
import sys

def load_data():
    lInf = []
    f=open('Dante/inferno.txt', encoding="ISO-8859-1")

    for line in f:
        lInf.append(line.strip())
    f.close()

    lPur = []
    f=open('Dante/purgatorio.txt', encoding="ISO-8859-1")

    for line in f:
        lPur.append(line.strip())
    f.close()

    lPar = []
    f=open('Dante/paradiso.txt', encoding="ISO-8859-1")

    for line in f:
        lPar.append(line.strip())
    f.close()
    return lInf, lPur, lPar

def split_data(l, n):
    lTrain, lTest = [], []
    for i in range(len(l)):
        if i % n == 0:
            lTest.append(l[i])
        else:
            lTrain.append(l[i])
            
    return lTrain, lTest

lInf, lPur, lPar = load_data()

lInf_train, lInf_evaluation = split_data(lInf, 4)
lPur_train, lPur_evaluation = split_data(lPur, 4)
lPar_train, lPar_evaluation = split_data(lPar, 4)

print(lInf_train)
```

```python
def concat_tercets(tercets):
    """
    Concatenate a list of words into a single string.
    """
    return ' '.join(tercets)

def split_words(text):
    """
    Split a string into a list of words.
    """
    return text.split()

def build_symbol_table(d):
    """
    Build a symbol table from a list of words.
    """
    return sorted(list(set(d)))

def lookup(symbol_table, word):
    """
    Lookup a word in the symbol table using binary search.
    """
    low, high = 0, len(symbol_table) - 1
    while low <= high:
        mid = (low + high) // 2
        if symbol_table[mid] == word:
            return mid
        elif symbol_table[mid] < word:
            low = mid + 1
        else:
            high = mid - 1
    return -1

def count_words(symbol_table, d):
    counts = np.zeros(len(symbol_table), dtype=int)
    for word in d:
        index = lookup(symbol_table, word)
        if index != -1:
            counts[index] += 1
    return counts

words = split_words(concat_tercets(lInf_train)) + split_words(concat_tercets(lPur_train)) + split_words(concat_tercets(lPar_train))
symbol_table = build_symbol_table(words)
print(symbol_table)
word_counts_inf = count_words(symbol_table, split_words(concat_tercets(lInf_train)))
word_counts_pur = count_words(symbol_table, split_words(concat_tercets(lPur_train)))
word_counts_par = count_words(symbol_table, split_words(concat_tercets(lPar_train)))

word_freqs_inf = (word_counts_inf + 0.001) / np.sum(word_counts_inf)
word_freqs_pur = (word_counts_pur + 0.001) / np.sum(word_counts_pur)
word_freqs_par = (word_counts_par + 0.001) / np.sum(word_counts_par)
```

```python
def predict_cantica(x):
    """
    Predict the cantica of a given text.
    """
    # Split the text into words
    words = split_words(x)
    word_indices = []
    for word in words:
        lookup_index = lookup(symbol_table, word)
        if lookup_index != -1:
            word_indices.append(lookup_index)

    llinf = np.log(word_freqs_inf[word_indices]).sum()
    llpur = np.log(word_freqs_pur[word_indices]).sum()
    llpar = np.log(word_freqs_par[word_indices]).sum()

    return np.array([llinf, llpur, llpar])



S = np.zeros((3, len(lInf_evaluation) + len(lPur_evaluation) + len(lPar_evaluation)), dtype=float)
D = lInf_evaluation + lPur_evaluation + lPar_evaluation
L = np.array([0] * len(lInf_evaluation) + [1] * len(lPur_evaluation) + [2] * len(lPar_evaluation))


for i, x in enumerate(D):
    S[:, i] = predict_cantica(x)
    
# Get the predicted class
predicted_class = np.argmax(S, axis=0)

# Calculate the accuracy split for each cantica
accuracy = np.zeros(3, dtype=float)
for i in range(3):
    accuracy[i] = np.sum(predicted_class[L == i] == i) / np.sum(L == i)
print("Accuracy for Inferno: ", accuracy[0])
print("Accuracy for Purgatorio: ", accuracy[1])
print("Accuracy for Paradiso: ", accuracy[2])

# Compute prediction between pairs of canticas
pairs = [(0, 2), (0, 1), (1, 2)]
for cantica0, cantica1  in pairs:
    print(f"Prediction between {cantica0} and {cantica1}")
    elements_of_these_canticas = (L == cantica0) | (L == cantica1)
    predicted_class = S[cantica0, elements_of_these_canticas] - S[cantica1, elements_of_these_canticas] > 0
    expected_class = (L[elements_of_these_canticas] == cantica0)
    accuracy = np.sum(predicted_class == expected_class) / len(expected_class)
    print(f"Accuracy", accuracy)
    
```
